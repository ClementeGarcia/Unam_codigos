```python
import numpy as np
import matplotlib.pyplot as plt

# Definir rangos
x = np.linspace(0, 2000, 1000)

# Restricciones
y1 = (2000 - x)                   # x + y = 2000
y2 = x / 2                         # x = 2y
x_500 = np.ones(1000) * 500        # x >= 500
y_500 = np.ones(1000) * 500        # y >= 500

# Graficar
plt.plot(x, y1, label='x + y = 2000')
plt.plot(x, y2, label='x = 2y')
plt.axvline(x=500, color='r', linestyle='--', label='x ≥ 500')
plt.axhline(y=500, color='g', linestyle='--', label='y ≥ 500')

# Punto de intersección (solución)
plt.plot(4000/3, 2000/3, 'ro', label='Solución óptima')

# Configuración
plt.xlim(0, 2000)
plt.ylim(0, 2000)
plt.xlabel('x₁')
plt.ylabel('x₂')
plt.legend()
plt.grid()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Configuración
plt.figure(figsize=(10, 8))
plt.xlabel('x₁', fontsize=12)
plt.ylabel('x₂', fontsize=12)
plt.title('Región Factible: $x_1 + x_2 \leq 2000$', fontsize=15)
plt.grid(True, linestyle='--', alpha=0.7)

# Rangos
x = np.linspace(0, 2000, 1000)

# Restricciones
y1 = 2000 - x                     # x₁ + x₂ ≤ 2000
y2 = x / 2                        # x₁ = 2x₂

# Graficar restricciones
plt.plot(x, y1, 'b-', label='$x_1 + x_2 \leq 2000$', linewidth=2)
plt.plot(x, y2, 'm-', label='$x_1 = 2x_2$', linewidth=2)
plt.axvline(x=500, color='r', linestyle='--', label='$x_1 \geq 500$')
plt.axhline(y=500, color='g', linestyle='--', label='$x_2 \geq 500$')

# Puntos extremos de la región factible
x_values = [1000, 1500]
y_values = [500, 500]



# Puntos de interés
plt.plot(1000, 500, 'ko', markersize=8, label='Punto extremo (1000, 500)')
plt.plot(4000/3, 2000/3, 'ro', markersize=8, label='Punto extremo (1333.33, 666.67)')

# Ajustes
plt.xlim(0, 2000)
plt.ylim(0, 2000)
plt.legend(loc='upper right')
plt.show()
```
